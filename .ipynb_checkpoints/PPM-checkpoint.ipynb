```python
import pandas as pd
```

# Phase 1 : Import, Drop Columns, Add Columns

## Read the file

### Main file

```python
df = pd.read_excel("./PPM.xlsx")
df.head()
```

### Reference file

```python
dfRef = pd.read_excel("./Ref.xlsx")
dfRef.head()
```

## Drop not important columns

```python
df.drop(['Flag'], axis=1)
df.drop(['Priority'], axis=1)
df.head()
```

## Add columns

### Line Column

```python
def getInput(txtFrom) :
    if "RED" in txtFrom :
        return "Red Line"
    elif "GREEN" in txtFrom :
        return "Green Line"
    elif "GOLD" in txtFrom :
        return "Gold Line"
    else : return "Blue Line"

df['Line'] = df['Location'].apply(lambda x: getInput(x))
df.head()
```

### Station Column

```python
def getInput(listOfRef, txtFrom) :
    #listOfRef = dfRef items
    #txtFrom = df current index

    for curRef in listOfRef :
        if curRef in txtFrom :
            return curRef

df['Station'] = df['Location'].apply(lambda x: getInput(dfRef['Code'], x))
df.head()
```
